```python
# Convert color channels from OpenCV to Matplotlib for display purposes
# b/c cv2.imshow doesn't work in Jupyter Notebooks. OpenCV uses BGR: Matplotlib uses RGB

b,g,r = cv2.split(img)
rgb_img = cv2.merge([r,g,b])
    
# Display image using matplotlib:

#Creates plot of original image, resized for visibility here.
plt.rcParams["figure.figsize"] = (20,15)

plt.imshow(rgb_img)
plt.title('PMTs')
plt.show()
```

```python
# Load image as grayscale
gray = imread('C:/Users/JAZ/github/Starfitter21/0MDTAx4/20180920_00/20180920_00_0_25_0000.jpg', cv2.IMREAD_GRAYSCALE)

#Starblob test image
#gray = imread('C:/Users/JAZ/github/Starfitter21/0MDTAx4/20180920_00/20180920_00_0_25_0047.jpg', cv2.IMREAD_GRAYSCALE)

#Creates plot of gray image, resized for visibility here.
plt.rcParams["figure.figsize"] = (20,15)

plt.imshow(gray, cmap='gray')
plt.title('Gray PMTs')
```

```python
# Use Laplacian of Gaussian (LoG) to find PMT edges

# Apply Gaussian Blur
blur = cv2.GaussianBlur(gray,(3,3),0)

#Creates plot of blurred image, resized for visibility here.
plt.rcParams["figure.figsize"] = (20,15)

plt.imshow(blur, cmap='gray')
plt.title('Blurred PMTs')
```
